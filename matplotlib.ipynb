*function to label each point, warning : the x-axis must be numeric, not str*

```python
def add_labels(dataframe, y_column, label_column, x_offset=0.3):
    """
    Add labels for a scatter plot
    
    Parameters:
    - dataframe
    - y_column : can be str
    - label_column : str = name of the column to retrieve texts for labels
    - x_offset: float, horizontal offset (d√©calage), default: 0.3
    """
    x_positions = range(len(dataframe))  # Must be numeric
    y_values = dataframe[y_column]       
    labels = dataframe[label_column]     
    
    for x, y, label in zip(x_positions, y_values, labels):
        plt.text(x + x_offset, y, label,
                 va='center',      
                 ha='left',        
                 fontsize=9) # plt.text = write label
```
