1. What is the average duration of top songs prior to 1968? After 1968?

```python
import pandas as pd
df = pd.read_csv("top-song-durations.csv")

# Select songs before 1968
before_1968 = df.query("year <= 1968")

# Split and convert
split_duration = before_1968["duration"].str.split(":", expand=True).astype("int")
# Create columns in the df
before_1968[["h", "m", "s"]] = split_duration
# Sum seconds
before_1968["duration_seconds"] = before_1968.eval("h * 3600 + m * 60 + s")
mean_before1968 = before_1968["duration_seconds"].mean()

# And same for after 1968
after_1968 = df.query("year >= 1968")
split_duration2 = after_1968['duration'].str.split(':', expand=True).astype('int')
after_1968[["h", "m", "s"]] = split_duration2
after_1968["duration_seconds2"] = after_1968.eval("h*3600 + m*60 + s")
mean_after1968 = after_1968["duration_seconds2"].mean()
mean_after1968
```
