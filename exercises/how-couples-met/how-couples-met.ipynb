1. Create a horizontal bar plot for a specific decade, displaying the proportion of people who met through each method (e.g., online, through friends)

```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('how-couples-met.csv')

# Select the 2000 decade
df_2 = df.query('decade == 2000')
df_2 = df_2.set_index('decade')
df_2.plot.bar()

# Configuration
ax = plt.gca()
ax.tick_params(axis='x', length=0)
ax.set_xticklabels([])
ax.set_xlabel('')
plt.ylabel('percentage (%)')

```

2. Create a set of subplots featuring horizontal bar plots for two different decades of your choosing, with each subplot showing the proportion of people who met through each method.

```python
# Dataframes for 1970 + transposition (invert y-axis and x-axis)
df_2000 = df.query('decade == 2000').set_index('decade').T
df_1970 = df.query('decade == 1970').set_index('decade').T

# Rename column for percentage
df_2000.rename(columns={2000:'percentage'})
df_1970.rename(columns={1970:'percentage'})

# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(15, 8))
axes[0].bar(df_2000.index, df_2000.iloc[:,0]) # .iloc[:,0] = select by position (not by name), ":" means all lines, "0" means for the first position
axes[0].set_title('How people met in 2000')
axes[1].bar(df_1970.index, df_1970.iloc[:,0], color='orange')
axes[1].set_title('How people met in 1970')
```
