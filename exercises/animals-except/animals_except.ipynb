1. Convert your findings into a research report examining how British menageries influenced the word frequency for different animal words.

```python
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.read_csv('animal-word-trends-menageries.csv')

animals = df['word'].unique()

plt.figure(figsize=(10,7))
for animal in animals:
    animal_df = df.query('word == @animal')
    plt.plot(animal_df['year'], animal_df['frequency'], label=animal)
plt.legend()
plt.axvspan(1750, 1835, color='magenta', alpha=0.1)

# Reporting most cited
menagerie = df.query('year <= 1835 and year >= 1750')
most_cited = menagerie.sort_values('frequency', ascending = False)
print(f'The animal most cited : {most_cited.values[0][1]}')

# Less cited
less_cited = menagerie.sort_values('frequency')
print(f'The animal less cited : {less_cited.values[0][1]}')

# Groups of animals less cited
group = less_cited.head(5)
animals_names = group['word'].reset_index(drop=True)
list_animals = []
for name in animals_names:
    if name not in list_animals:
        list_animals.append(name)
result = ', '.join(list_animals)
print(f'Animal less cited : {result}.')
    
```
