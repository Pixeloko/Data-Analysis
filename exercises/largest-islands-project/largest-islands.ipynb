# World's largest islands

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('largest-islands.csv')
df.head()
```

### Project Ideas

- What are the 10 largest islands in the tropics?

- What are the largest islands in each `region`?

- Create a line graph with `area` on the y-axis and `rank` on the x-axis. The data should be ordered by `rank`, from largest to smallest.

- What islands are composed of multiple countries?
	- Hint: try `countries.str.contains(",")`

```python
# Largest islands
largests = df.nlargest(10, 'area')

# Largest islands in each regions
regions_groups = df.loc[df.groupby('region')
                        ['area'].idxmax()
]

# Create a plot y=area, x=rank sorted by rank
regions_sorted = regions_groups.sort_values('rank', ascending=False)
plt.plot(regions_sorted['rank'], regions_sorted['area'])
plt.xlabel('rank')
plt.ylabel('area')

# Islands composed of multiple countries
multiple_countries = df.query('countries.str.contains(",")')
```
