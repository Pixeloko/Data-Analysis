```python
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.read_csv('animal-word-trends-intro.csv')
```

1. Are "cats" or "dogs" more popular in publications? Graph the word trends for both animals to find out!

```python
def word_trend(animal):
    animal_df = df.query('word == @animal') 
    plt.plot(animal_df['year'], animal_df['frequency'], label=animal)

word_trend('cat')
word_trend('dog')
plt.title(f'Animal word trend in printing materials')
plt.legend()
```
