## Project extensions

Convert your findings into a research report examining the multi-year repeating patterns of cicadas.

```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('animal-word-trends-cicada.csv')
broods = pd.read_csv('cicada-brood-mentions-print.csv')
plt.plot(df['year'], df['frequency'])

# Report
avg_before_1900 = df.query('year < 1900')['frequency'].mean()
avg_after_1900 = df.query('year > 1900')['frequency'].mean()
diff_freq = ((avg_after_1900 - avg_before_1900)/avg_before_1900) * 100

print(f'-----------REPORT-----------\n* Frequency of cicada before 1900 : {avg_before_1900:.2f}')
print(f'* Frequency of cicada before 1900 : {avg_after_1900:.2f}')
print(f'* Difference : {diff_freq:.2f} %')

# Concerning broods
most_mentioned = broods.sort_values('mentions', ascending=False)
print(f'* The most mentionned brood : Brood {most_mentioned.values[0][0]}')
print(f'* With {most_mentioned.values[0][1]} mentions')

```
