```python
# 1. Which bird has the fewest neck vertebra
import pandas as pd

fewer_birds_vertebrae = pd.read_csv('bird-neck-bones.csv').query('neck_vertebrae == neck_vertebrae.min()')
print(fewer_birds_vertebrae)
```

```python
# 2. Do humans have more bones in the arms or legs?
humans = pd.read_csv('adult-human-skeleton.csv')
# Filter the rows where the region is arm
arm_region = humans.query('region == "arm"')
# Count 
print(f"Number of arms' bones {arm_region['region'].value_counts()}\n")
# Do the same for leg
leg_region = humans.query('region == "leg"')
print(f"Number of legs' bones {leg_region['region'].value_counts()}")
```

```python
# 3. How many ribs do humans have?
# Filter the rows where the name contain rib
import pandas as pd
humans = pd.read_csv('adult-human-skeleton.csv')

# str.contains() used on series
# case = False means case insensitive
filtered_humans = humans[humans['name'].str.contains('rib', case=False, na=False)]
print(filtered_humans)

# Count the number of rows
filtered_humans.shape[0]
```
