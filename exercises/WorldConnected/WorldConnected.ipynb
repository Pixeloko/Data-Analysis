1. Determine the year when more than 50% of the population in a continent of your choice first gained internet access.


```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the two files
EuropePop = pd.read_csv('extension-historical-population-by-continent.csv').query('continent == "Europe"')
InternetEuropeanPop = pd.read_csv('extension-internet-users-by-continent.csv').query('continent == "Europe"')

# Merge
dataEU = EuropePop.merge(InternetEuropeanPop, on="year", how="left").dropna()

# Create column for percentage of connected population
dataEU["percent"] = dataEU.eval(dataEU["internet_users"] / dataEU["population"] * 100)

# Query only more than 50 % year
MoreThanHalf = dataEU.query('percent >= 50')

print(MoreThanHalf.round(2).head(1))
```
