Below is a proposed extension idea. 

The file `us-daily-passengers.csv` contains data on the daily number of passengers flying in the United States throughout 2023.

Create a graph that shows the average number of passengers flying per day of the week. Then, compare this graph to your graph of the percentage of flights delayed by day of week. What patterns do you see?

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('us-daily-passengers.csv')

# Convert date into datetime type
data['date'] = pd.to_datetime(data["date"])

# Create a new df containing the needed elements
flights = data[['date', 'num_passengers']].copy()

# Add day name column
flights['day'] = flights['date'].dt.strftime('%a')

# Group by day and display the average number of passengers
daily_avg = flights.groupby(['day'])['num_passengers'].mean()

# Rearrange the index 
new_index_order = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
daily_avg = daily_avg.reindex(new_index_order)

# Create the plot
plt.bar(daily_avg.index, daily_avg)
```
